```python
import yfinance as yf
import os
import pandas as pd

# company tickers
companies = ["OGDC.KA", "PPL.KA", "HBL.KA", "MCB.KA", "LUCK.KA", "DGKC.KA"]

# create folder
save_path = "data_raw"
os.makedirs(save_path, exist_ok=True)

# download company data
for ticker in companies:
    print(f"Downloading {ticker} ...")
    data = yf.download(ticker, start="2020-01-01", end="2025-11-28")

    file_name = f"{ticker.replace('.KA', '')}.csv"
    data.to_csv(f"{save_path}/{file_name}")
    print(f"Saved: {file_name}")

# download index
'''print("Downloading KSE-100 index (^KSEI) ...")
index = yf.download("^KSEI", start="2020-01-01", end="2025-11-28")
index.to_csv(f"{save_path}/KSE100.csv")
print("Saved: KSE100.csv")'''

print("All downloads completed.")
```
